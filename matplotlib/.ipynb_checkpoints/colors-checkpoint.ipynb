## simples color names

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Sample data
x = np.arange(10)
y = x**2

fig, ax = plt.subplots()

# Bar chart
ax.bar(x, y, label="Bar Chart", alpha=0.6)

# Line plots with different colors
ax.plot(x, y, color='red', label="Red Line")
ax.plot(x, y, color='blue', label="Blue Line")
ax.plot(x, y, color='green', label="Green Line")

ax.set_title("Bar + Multiple Line Plots")
ax.legend()

plt.show()
```

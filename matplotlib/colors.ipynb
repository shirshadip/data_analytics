## simples color names

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(7,4))

ax.plot(x, y1, color='blue', linestyle='--', marker='o',
        linewidth=2, markersize=5, label='sin(x)')
ax.plot(x, y2, color='red', linestyle='-', marker='s',
        linewidth=2, markersize=5, label='cos(x)')

ax.set_title("Styled Line Plot", fontsize=16, fontweight='bold')
ax.set_xlabel("x-axis", fontsize=12)
ax.set_ylabel("y-axis", fontsize=12)

ax.legend()
ax.grid(True)

plt.show()
```

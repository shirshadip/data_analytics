## 1.line plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()

x = np.linspace(0,10,100)
# print (x)
y = np.sin(x)
# print (y)
ax.plot(x,y)
ax.set_title("line plot")
plt.show()
```

## 2.scatter plot

```python
fig , ax = plt.subplots()
x =np.random.rand(50)
y = np.random.rand(50)
ax.scatter(x,y,marker="o")
ax.set_title("scatter plot")
plt.show()
```

## 3.bar chart

```python
fig ,ax = plt.subplots()
categories = ['A','B','C','D']
values =[10,23,15,8]
ax.bar(categories,values)
ax.set_title("Bar plot")
plt.show()
```

## 4. Histogram
### used for .
i. distribution

ii. understanding how data is spread

```python
fig ,ax = plt.subplots()
data=np.random.randn(2000)

ax.hist(data,bins=100)
ax.set_title("Histogram plot")
plt.show()
```

###  Divides data into bins
###  Shows how many points fall in each bin

## 5. Pie chart
### used for.
i. percentage
ii. proportions

```python
fig ,ax = plt.subplots()
sizes=[30,25,20,25]
labels=['Python','Java','C','C++']

ax.pie(sizes,labels=labels, autopct="%1.1f%%")
ax.set_title("Pie chart")
plt.show()
```

## 6.Box plot
Used for:

statistical summary

median, quartiles, outliers

comparing distributions


```python
fig,ax=plt.subplots()
data=[np.random.randn(100),np.random.randn(100)+1]
ax.boxplot(data)
ax.set_title("Box plot")
ax.set_xticklabels(['Group 1','Group 2'])
plt.show()
```

 Shows median
 Box = Q1 to Q3
 Whiskers = range
 Dots = outliers

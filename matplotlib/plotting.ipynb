# mini example


```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,200)
y =np.sin(x)
# # print (x)
# print (y)
plt.figure(figsize=(12,5))
plt.plot(x,y)
plt.title("sin wave")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

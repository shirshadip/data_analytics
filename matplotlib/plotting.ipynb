# mini example


```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,200)
y =np.sin(x)
# # print (x)
# print (y)
plt.figure(figsize=(12,5))
plt.plot(x,y)
plt.title("sin wave")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

# same plot 
different method 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6,4))  # create Figure and Axes objects
ax.plot(x, y, label='sin(x)')
ax.set_title("Sine wave (OO API)")
ax.set_xlabel("x (radians)")
ax.set_ylabel("sin(x)")
ax.legend()
ax.grid(True)

fig.tight_layout()
fig.savefig("sine_wave.png", dpi=150)  # export to file
plt.show()
```

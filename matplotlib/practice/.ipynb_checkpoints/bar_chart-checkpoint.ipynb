### 3) Create a bar chart

Categories = ["Apple", "Mango", "Banana"]
Values = [50, 70, 30]

```python
import matplotlib.pyplot as plt

cat = ["Apple", "Mango", "Banana"]
values = [50,70,30]
fig,ax = plt.subplots()
ax.bar(cat,values)
plt.xticks(rotation=90)
ax.set_title("Sales report")
ax.set_xlabel("Categories")
ax.set_ylabel("Sales")
plt.show()
```

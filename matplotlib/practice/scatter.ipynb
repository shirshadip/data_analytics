# 1) Create a scatter plot of 100 points where

x = random numbers between 0 and 1

y = xÂ² + some random noise

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
x = np.random.rand(100)
noise= np.random.randn(100)*0.1
y =x**2+noise
ax.scatter(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()

# print (y)
```

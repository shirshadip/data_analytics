Create a plot with these rules:

sin(x) ‚Üí green, dashed (--), circle markers (o), linewidth=2, markersize=7

cos(x) ‚Üí purple, dotted (:), square markers (s), linewidth=2, markersize=7

legend position ‚Üí upper left

grid ‚Üí ON

title ‚Üí ‚ÄúStyled Trigonometric Plot‚Äù

üëâ You write the code.

```python
from cProfile import label

import numpy as np
import matplotlib.pyplot as plt

```

```python
x = np.linspace(0,10,200)
y1=np.sin(x)
y2=np.cos(x)

fig , ax = plt.subplots()
#sin(x)
ax.plot(
   x, y1,
    color='green',
    linestyle='--',
    marker='o',
    linewidth=2,
    markersize=7,
    label='sin(x)'


)
#cos (x)
ax.plot(
     x, y2,
    color='purple',
    linestyle=':',
    marker='s',
    linewidth=2,
    markersize=7,
    label='cos(x)'
)
ax.set_title("styled trigonometric expression")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(loc="upper left")
ax.grid(True)
plt.show()
```

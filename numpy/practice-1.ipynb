## Weather data analytics

a numpy array of 365 days daily temparature 

```python
import numpy as np 
import matplotlib.pyplot as plt
temp = np.random.randint(15,45,size=365)
print ("the temparature of 365 days is:\n",temp)
```

### lets find the hottest temperature and the coolest tempetrature

```python
hot = temp.max()
cool=temp.min()
print ("the hottest temparature of a day:",hot,"degree celcius in day no.",temp.argmax())
print ("the coolest temparature of a day:",cool,"degree celcius in day no.",temp.argmin())
```

## average temparature

```python
avg=0
index=0
for days in temp:
    avg = temp.mean()
    # print("the avg temp is:", avg)
    index+=1
print ("the avg tem is :",round(avg,2),f"degree celcius of {index} days")
```

## how many days were upto 35 degree celcius

```python
t5up=np.sum(temp>=35)
print(f"{t5up} days were upto 35 degree celcius")
```

# visualized data

```python
plt.figure(figsize=(15,4))
plt.plot(temp)
plt.xlabel("Day of the year")
plt.ylabel("Temperature (C)")
plt.title("weather Data Visualization")
plt.show()
```
